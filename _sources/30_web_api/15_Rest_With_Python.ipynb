## Rest with Python

```python
# first we need to import some modules
# utility to read json format
import json as js
# urllib2 is a module for fetching URL (Uniform Resource Locator) data
import urllib.request
# import pandas
import pandas as pd
```

```python
# following this example: http://stackoverflow.com/questions/13921910/python-urllib2-receive-json-response-from-url

request_string="http://maps.googleapis.com/maps/api/elevation/json?locations=42,73&sensor=false"    
response=urllib.request.urlopen(request_string)
print(type(response))
data=js.load(response)
print(type(data))
print(data)
```

```python
request_string="http://api.fda.gov/drug/event.json?search=patient.drug.medicinalproduct:aspirin&count=patient.reaction.reactionmeddrapt.exact"
response=urllib.request.urlopen(request_string)
print(type(response))
data=js.load(response)
print(type(data))

print(js.dumps(data, indent=4))
```
