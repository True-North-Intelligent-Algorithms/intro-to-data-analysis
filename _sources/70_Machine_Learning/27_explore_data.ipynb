## Explore Data

In this notebook we open up a well known dataset which contains health metrics and diabetes status for a group of Prima Indians.  For more details see [here](https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database).

The objective of this notebook is to open and explore the data.  We generate different types of plots and try to see if we can find health metrics that predict diabetes status. 

## Open Data with Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

diabetes=pd.read_csv('indians-diabetes.csv')

print (diabetes.shape)
print (diabetes.columns)
diabetes.head()
```

## Define a function to make a 2 class scatterplot.

This function generates a scatter plot using two features.  The points on the plot are given different colors for the 'No diabetes' and 'Diabetes' classes.  If the two features are good predictors we should see two distinct clusters.  If the two features are not good predictors we should see only one cluster.  In real life (and for the diabetes dataset) the situation is somewhat in between.  

```python
def plot_variables(x_variable,y_variable):
    ax=diabetes[diabetes['class']==0].plot.scatter(x=x_variable,y=y_variable,color='blue',label='No diabetes')
    diabetes[diabetes['class']==1].plot.scatter(x=x_variable,y=y_variable,color='red',label='Diabetes',ax=ax)
    return ax

plot_variables('age','bp')
```

# Define a function to make a two class histogram 

This plot can be used to see if one feature alone is a good predictor.  If the feature can predict the class we should see two distinct peaks in the combined histogram.  If the feature is not correlated with the class the histograms should be indistinguishable.   In real life (and for this dataset) the situation is somewhat in between.

```python
def plot_hists(x_variable):
    ax=diabetes[diabetes['class']==0][x_variable].plot.hist(color='blue',alpha=0.4, label='No diabetes')
    diabetes[diabetes['class']==1][x_variable].plot.hist(color='red',alpha=0.4,label='Diabetes',ax=ax)
    plt.legend()
    return ax

plot_hists('bmi')
```

```python
# box plot
diabetes.boxplot(column='bmi',by='class')
```
