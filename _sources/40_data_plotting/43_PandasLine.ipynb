## Using Pandas for line plot

All Pandas plotting workflows start with the same two line.  The first one tells the notebook to plot inline (in the notebook instead of a popup), and the second imports pandas and assigns the module to the shortcut 'pd'. 

````python
%matplotlib inline
import pandas as pd
````

Plotting with Pandas is relatively simple.  First we open the data and get a dataframe using the standard read_csv function

````python
msft=pd.read_csv('microsoft.csv')
print(type(msft))
print (msft.head())
````

And now we can plot by calling functions directly from the dataframe...

````python
#msft.Data=msft.Date.astype('datetime64')
msft.set_index('Date', inplace=True)
msft=msft.reindex(index=msft.index[::-1])
````

The ```plot()``` fanction returns a ```axis``` object.  This is useful if we want to set the title, labels and other properties of the plot.

````python
axes=msft.Close.plot(figsize = (15,10))
print(type(axes))
axes.set_title("test")
axes.set_xlabel("time")
axes.set_ylabel("Closing price")
````
